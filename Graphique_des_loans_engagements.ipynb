```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
path='/content/Récapitulatif des Souscriptions Loans.xlsx'
sheet_name='Récap'
df = pd.read_excel(path, sheet_name=sheet_name,usecols = 'A:V')
df= df.dropna(subset=['Nom'])
```

```python

df2=df.loc[:,['Nom','Engagement']]
df2['engagement_pourc']=df['Engagement']/df['Engagement'].sum()*100
df2=df2.sort_values(by=['engagement_pourc'],ascending=False)
df2=df2.reset_index(drop=True)
```

```python
fig = plt.figure()

plt.figure(figsize=(12,8))
g1=sns.barplot(x=df2['Nom'], y=df2['engagement_pourc'], palette="Blues_r")
plt.xlabel([], fontsize=15, color='#2980b9')
plt.ylabel('Engagements %\n', fontsize=15, color='#2980b9')
#plt.title("Répartition des engagements Loans\n", fontsize=18, color='#2980b9')
plt.xticks(rotation= 80)
g1.set(xlabel=None)
sns.despine(bottom = False, left =False)

for index, row in df2.iterrows():
    g1.text(row.name,row.engagement_pourc, round(row.engagement_pourc,1), color='black', ha="center")


plt.tight_layout()
```

```python
def graphloans(num_soc):
  path='/content/Récapitulatif des Souscriptions Loans.xlsx'
  sheet_name='Récap'
  df = pd.read_excel(path, sheet_name=sheet_name,usecols = 'A:V')
  df= df.dropna(subset=['Nom'])
  nom_col_engagement=['Engagement','Engagement.1','Engagement.2']


  df2=df.loc[:,['Nom',nom_col_engagement[num_soc-1]]]
  df2=df2.dropna()
  df2['engagement_pourc']=df2[nom_col_engagement[num_soc-1]]/df2[nom_col_engagement[num_soc-1]].sum()*100
  df2=df2.sort_values(by=['engagement_pourc'],ascending=False)
  df2=df2.reset_index(drop=True)

  fig = plt.figure()

  plt.figure(figsize=(12,8))
  g1=sns.barplot(x=df2['Nom'], y=df2['engagement_pourc'], palette="Blues_r")
  plt.xlabel([], fontsize=15, color='#2980b9')
  plt.ylabel('Engagements %\n', fontsize=15, color='#2980b9')
  #plt.title("Répartition des engagements Loans\n", fontsize=18, color='#2980b9')
  plt.xticks(rotation= 80)
  g1.set(xlabel=None)
  sns.despine(bottom = False, left =False)

  for index, row in df2.iterrows():
      g1.text(row.name,row.engagement_pourc, round(row.engagement_pourc,1), color='black', ha="center")


  plt.tight_layout()
```

```python
graphloans(1)
```
