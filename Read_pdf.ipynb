````python
# importing required modules
import PyPDF3
import pandas as pd
from io import StringIO

 
# creating a pdf file object
pdfFileObj = open('/content/liste des fonds boursorama.pdf', 'rb')
 
# creating a pdf reader object
pdfReader = PyPDF3.PdfFileReader(pdfFileObj)
 
#printing number of pages in pdf file
print("PDF number of page:{}".format(pdfReader.numPages))
num_pages=pdfReader.numPages
 
# creating a page object
data=""
for page in range(0,num_pages):
    pageObj = pdfReader.getPage(page)
    pageObjtext=pageObj.extractText()
    data+=pageObjtext


 
# extracting text from page
#print(pageObj.extractText())

#print(type(data))
# closing the pdf file object
pdfFileObj.close()

StringData = StringIO(data).read()
#print(StringData.read())
#print(StringData)
#df = pd.read_csv(StringData, sep ="\n")
#df=df[17:42]
#print(df.head(50))
#population_data = pd.DataFrame(columns=["Libellé", "Code ISIN", "Nature juridique", "Société de gestion", "Performance brute de l'actif N-1","Frais de gestion de l’actif"])

````
